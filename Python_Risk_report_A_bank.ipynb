```python
from googletrans import Translator

# Функция для перевода текста
def translate_text(text):
    try:
        translation = translator.translate(text, src='en', dest='ru')
        return translation.text
    except Exception as e:
        print(f"Error translating '{text}': {e}")
        return text  # Возвращаем оригинальный текст в случае ошибки

# Применяем функцию перевода к столбцу
df12['russian_text'] = df12['DESCRIPTION'].apply(translate_text)

# Выводим результат
df12.head(2)
```

```python
import re

# Функция для разбивки строки на слова
def split_camel_case(text):
    if isinstance(text, str):  # Проверяем, что текст является строкой
        return re.findall(r'[A-Z][a-z]*|[a-z]+', text)
    else:
        return []  # Возвращаем пустой список для нестроковых значений

# Применяем функцию к столбцу 'NAME'
df12['Split_Category'] = df12['NAME'].apply(split_camel_case)

df13 = df12[['NAME', 'Split_Category', 'DESCRIPTION', 'russian_text']]

# Преобразуем список в строку
df13['Split_Category_String'] = df13['Split_Category'].apply(lambda x: ' '.join(x))
df13.head(2)
```

### 3. Определите самые прибыльные сегменты (не менее 2-х). Почему вы так думаете?

```python
# Функция для определения диапазона
def assign_range(value):
    if 0 <= value < 0.2:
        return '0-0.2'
    elif 0.2 <= value < 0.4:
        return '0.2-0.4'
    elif 0.4 <= value < 0.6:
        return '0.4-0.6'
    elif 0.6 <= value < 0.8:
        return '0.6-0.8'
    elif 0.8 <= value < 1:
        return '0.8-1'
    else:
        return None

# Применяем функцию к столбцу и создаем новый столбец
df['ProbabilityOfDefault_range'] = df['ProbabilityOfDefault'].apply(assign_range)
```

```python
df_full = df.groupby(['County', 'City', 'Age_range','LoanDuration', 'CreditScoreEeMini', 'ProbabilityOfDefault_range'], as_index=False) \
    .agg({'LoanId': 'nunique', 'Amount': 'sum', 'Interest': 'mean'}) \
    .sort_values(['Amount', 'Interest'], ascending= [False, False]) \
    .rename(columns={'LoanDuration': 'Продолжительность_займа',
                     'County': 'Страна',
                     'City': 'Город',
                     'LoanId': 'Количество_заемщиков',
                     'Age_range': 'Возрастная_категория',
                     'CreditScoreEeMini': 'Кредитный_рейтинг_EE_Mini',
                     'Amount': 'Сумма_займа',
                     'ProbabilityOfDefault_range': 'Вероятность_дефолта',
                     'Interest': '%_среднее'}) \
    .head(5)

score_map_8 = {1000: 'нет предыдущих проблем с платежами', 900: 'проблемы с платежами закончились 24-36 месяцев назад', 
               800: 'проблемы с платежами закончились 12-24 месяца назад', 700: 'проблемы с платежами закончились 6-12 месяца назад', 
               600: 'проблемы с оплатой завершены <6 месяцев назад', 500: 'проблемы с активной оплатой'}
score_map_9 = {'0-0.2': 'очень низкий', 
               '0.2-0.4': 'низкий', 
               '0.4-0.6': 'средний', 
               '0.6-0.8': 'средний высокий', 
               '0.8-1': 'очень высокий'}

df_full['Значение_риска_EE_Mini'] = df_full['Кредитный_рейтинг_EE_Mini'].map(score_map_8)
df_full['Значение_дефолта'] = df_full['Вероятность_дефолта'].map(score_map_9)

df_full['Средняя_сумма_займа'] = round(df_full.Сумма_займа / df_full.Количество_заемщиков, 0)
df_full['Прибыль'] = round(df_full.Сумма_займа * (df_full['%_среднее']/100), 2)
df_full['%_среднее'] = round(df_full['%_среднее'], 2)

df_full = df_full[['Страна', 'Город', 'Возрастная_категория', 'Продолжительность_займа', 
                   'Вероятность_дефолта', 'Значение_дефолта',
                   'Кредитный_рейтинг_EE_Mini', 'Значение_риска_EE_Mini', 'Количество_заемщиков', 
                   'Сумма_займа', 'Средняя_сумма_займа', '%_среднее', 'Прибыль']]
df_full = df_full.sort_values('Прибыль', ascending=False)
df_full
```
